```python
# Webpage URL
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

# Define the column names
col_names = ["sepal_length_in_cm",
            "sepal_width_in_cm",
            "petal_length_in_cm",
            "petal_width_in_cm",
            "class"]

# Read data from URL
iris_data = pd.read_csv(url, names=col_names)

iris_data.head()
```

```python
import seaborn as sns
```

```python
sns.pairplot(iris_data,hue='class')
```

```python
iris_data.plot(kind='scatter',x='sepal_length_in_cm', y='sepal_width_in_cm')
```

```python
sns.jointplot(x='sepal_length_in_cm', y='sepal_width_in_cm',data=iris_data, size=5)
```

```python
import matplotlib.pyplot as plt
```

```python
# use seaborn's FacetGrid to color the scatterplot by species
sns.FacetGrid(iris_data, hue = 'class') \
    .map(plt.scatter, 'sepal_length_in_cm','sepal_width_in_cm') \
    .add_legend()
```

```python
sns.boxplot(x='class', y='petal_length_in_cm', data=iris_data)
```

```python
ax = sns.boxplot(data=iris_data, x = 'class',y = 'petal_length_in_cm')
ax = sns.stripplot(data=iris_data, x='class', y='petal_length_in_cm', jitter=True, edgecolor='green')
```

```python
sns.violinplot(x='class',y='petal_length_in_cm', data=iris_data)
```

```python
sns.FacetGrid(iris_data, hue="class") \
   .map(sns.kdeplot, "petal_length_in_cm") \
   .add_legend()
```
